```python
data = pd.read_excel('APP滿意度調查 (回覆內容).xlsx',sheetname='表單回應 1')
```

```python
newdata = data.fillna(0)
d = newdata.values
```

```python
def ToMatrix(x):
    Y = []
    x = x[1:]
    length = len(x)
    W = np.matrix(np.zeros([length,length]))
    for i in range(length):
        if x[i] == 0.0:
            Y.append(np.zeros([1,length])[0].tolist())   
        else:
            a = []
            for j in range(length):
                if x[j] == 0.0 :
                    a.append(0)
                else :
                    a.append(x[j]-x[i])
                    W[i,j] = 0.05
            Y.append((a))
    return np.matrix(Y), W
```

```python
length = len(d[0])-1
A = np.matrix(np.zeros([length,length]))
W = np.matrix(np.zeros([length,length]))
for i in d:
    B,W1 = ToMatrix(i)
    A = A + B
    W = W + W1
```

```python
for i in range(length):
    for j in range(length):
        if W[i,j]!=0:
            A[i,j] = A[i,j]/W[i,j]
```

```python
def laplacian(Y):
     n = len(Y)
     x = -Y
     for i in range(n):
         x[i,i] = sum(Y[i,:])
     return x

def divergence(Y):
     n = len(Y)
     x = np.zeros([n,1])
     for i in range(n):
         x[i] = sum(Y[i,:])
     return x
```

```python
B = laplacian(np.array(W))
c = divergence(np.array(A))
```

```python
D = np.matrix(np.linalg.pinv(B))
```

```python
s = -D*c
```
